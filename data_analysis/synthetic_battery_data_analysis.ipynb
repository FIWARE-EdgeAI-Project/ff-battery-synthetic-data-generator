```python
# Battery Data Analysis

# This notebook analyzes the synthetic battery data generated by our simulator.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plot style
#plt.style.use('seaborn')
#sns.set_palette("deep")

# Load the data
data_dir = '../synthetic_datasets'
dataframes = {}

for filename in os.listdir(data_dir):
    if filename.endswith('.csv'):
        scenario = filename.replace('_battery_data.csv', '')
        df = pd.read_csv(os.path.join(data_dir, filename))
        df['Date'] = pd.to_datetime(df['Date'])
        dataframes[scenario] = df

print(f"Loaded {len(dataframes)} scenarios.")

# 1. SOH Over Time
def plot_soh_over_time(dataframes):
    plt.figure(figsize=(12, 6))
    for scenario, df in dataframes.items():
        plt.plot(df['Date'], df['SOH'], label=scenario)
    plt.xlabel('Date')
    plt.ylabel('State of Health (%)')
    plt.title('Battery State of Health Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_soh_over_time(dataframes)

# 2. SOH vs Cycles
def plot_soh_vs_cycles(dataframes):
    plt.figure(figsize=(12, 6))
    for scenario, df in dataframes.items():
        plt.scatter(df['Total_Cycles'], df['SOH'], label=scenario, alpha=0.5)
    plt.xlabel('Total Cycles')
    plt.ylabel('State of Health (%)')
    plt.title('Battery State of Health vs Total Cycles')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_soh_vs_cycles(dataframes)

# 3. SOH and Temperature
def plot_soh_and_temperature(dataframes):
    plt.figure(figsize=(12, 6))
    for scenario, df in dataframes.items():
        plt.scatter(df['Temperature'], df['SOH'], label=scenario, alpha=0.5)
    plt.xlabel('Temperature (°C)')
    plt.ylabel('State of Health (%)')
    plt.title('Battery State of Health vs Temperature')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_soh_and_temperature(dataframes)

# 4. SOH and Charging Behavior Influence
def plot_soh_and_charging_behavior(dataframes):
    plt.figure(figsize=(12, 6))
    for scenario, df in dataframes.items():
        charge_range = df['Charge_End'] - df['Charge_Start']
        plt.scatter(charge_range, df['SOH'], label=scenario, alpha=0.5)
    plt.xlabel('Charging Range (%)')
    plt.ylabel('State of Health (%)')
    plt.title('Battery State of Health vs Charging Range')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_soh_and_charging_behavior(dataframes)

# Conclusion
# Summarize your findings here. Discuss the patterns you observe in the plots and what they imply about battery degradation under different scenarios.
```
